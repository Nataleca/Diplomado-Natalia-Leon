## **3. Realice un _"describe"_ de todas las variables por año. Además de realizar un gráfico como boxplot con el que pueda comparar los resultados dados.**

```python
datos3=pd.read_csv(folder_path+"PRSA_data_2010.1.1-2014.12.31.csv",index_col=[1]).drop(['No'],axis=1)
datos3=datos3.dropna(subset=["pm2.5"])
display(datos3)
```

```python
# Se realiza el describe() para TODAS las variables
datos3.groupby('year').describe(include='all')
```

```python
# Seleccionamos solamente las variables cuantitativas
datos4=datos3.iloc[:,[3,4,5,6,8,9,10]]
display(datos4)
```

```python
datos4.groupby('year').describe()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
labels=['pm2.5','DEWP','TEMP','PRES','Iws','Is','Ir']
fig = plt.figure(figsize =(10, 7))
ax = fig.add_axes([0, 0, 1, 1])
bp = ax.boxplot(datos4,labels=labels,patch_artist = True)
ax.set_title('Box-Plot por Variable')

colors = ['#0000FF', '#00FF00', 
          '#FFFF00', '#FF00FF']
  
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
    
plt.show()
```

```python
# Ahora realizamos el Box-plot por año para la variable pm2.5

pm2_5=datos.loc[:,['year','pm2.5']].dropna(subset=["pm2.5"])
pm2_5.head()
pm2_5.boxplot(by='year')
plt.show()
```

```python
# Ahora realizamos el Box-plot por año para la variable DEWP

dewp=datos.dropna(subset=["pm2.5"])
dewp=dewp.loc[:,['year','DEWP']]
dewp.head()
dewp.boxplot(by='year')
plt.show()
```

```python
# Ahora realizamos el Box-plot por año para la variable TEMP

temp=datos.dropna(subset=["pm2.5"])
temp=temp.loc[:,['year','TEMP']]
temp.head()
temp.boxplot(by='year')
plt.show()
```

```python
# Ahora realizamos el Box-plot por año para la variable PRES

pres=datos.dropna(subset=["pm2.5"])
pres=pres.loc[:,['year','PRES']]
pres.head()
pres.boxplot(by='year')
plt.show()
```

```python
# Ahora realizamos el Box-plot por año para la variable Iws

iws=datos.dropna(subset=["pm2.5"])
iws=iws.loc[:,['year','Iws']]
iws.head()
iws.boxplot(by='year')
plt.show()
```

## **4. Realice un gráfico de dispersión de las variables "pm2.5" y "TEMP" con título, labels apropiados y color rojo para los datos. Describa: ¿Qué se puede inferir de dicho gráfico?**

```python
disp=datos.dropna(subset=["pm2.5"])
disp=disp.loc[:,['pm2.5','TEMP']]
plt.figure(figsize=(10, 6))
plt.plot(disp['pm2.5'],disp['TEMP'],'p',color='tab:red')
plt.title('Gráfico de Dispersión',fontdict={'fontsize':18})
plt.xlabel('pm2.5')
plt.ylabel('Temp')
plt.show()
```

* La mayor cantidad de datos es menor a 500 pm2.5.
* La varianza de los datos no es constante, de hecho, parece ser creciente
* No parece existir una relación "lineal" entre las variables pm2.5 y Temp


## **5. Calcule la probabilidad de encontrar en la Base de datos el valor de pm2.5 encima de 120**
