## <span style="color:#8F3A84"> **Relación Entre Variables** </span>

Hoy en día es de suma importancia el acceso a la educación superior, esto por supuesto, impulsa el desarrollo de un país y debería ser una prioridad gubernamental. Un profesional recién egresado, se enfrenta a múltiples cuestionamientos relacionados con su porvenir laboral e intelectual, y, en este contexto, cuando se decide continuar con estudios de postgrado, se desea asegurar el ingreso a un programa y universidad convenientes. _**¿Qué variables pueden aumentar la probabilidad de ser admitidos bajo las circunstancias deseadas?**_

Supongamos por ejemplo, que después de revisar distintas fuentes de información, se ha establecido que el dominio de una segunda lengua es un factor relevante en los procesos de admisión en programas de postgrado. En otras palabras,sospechamos que existe una **relación** entre la **probabilidad de ser admitido** y el **puntaje** obtenido en alguna prueba estandarizada de dominio de una segunda lengua. Lo que aún no hemos determinado, es _cómo es esa relación_ ni qué puede decirnos sobre los que debería hacerse para garantizar la admisión.

Introduzcamos ahora un poco de notación matemática:

* Llamemos $\pmb{x}$ al puntaje de la prueba de dominio de la segunda lengua.
* Denotemos con $\pmb{y}$ a la probabilidad de ser admitido en el programa de postgrado deseado.

Estamos suponiendo entonces, que algún cambio en el puntaje $x$, desencadena también un cambio en la probabilidad $y$. Desde el lenguaje matemático, esta relación entre las variables generalmente se representa con la siguiente expresión:

<span style="color:#033D74">$$\pmb{y=f(x)}$$</span>

Aquí, llamamos a $\pmb{y}$ **variable respuesta** o **dependiente**, mientras que $\pmb{x}$ se conoce como **covariable** o **variable independiente**. Existen muchas formas de relacionar estas variables, observa:

<figure>
  <img src="fig1.png" alt="Relaciones entre dos variables">
</figure>


En particular, cuando hablamos de una _**relación lineal**_ entre las variables, empleamos la notación


<span style="color:#033D74">$$\pmb{y=\beta_0+\beta_1x}$$</span>


Entendemos entonces que los cambios producidos en $y$ cuando varía $x$, pueden describirse a partir de la ecuación de una **recta** como la que aparece en la primera gráfica, y donde:

* $\pmb{\beta_0}$ representa el **intercepto** de la recta con el eje $y$, es decir, indica el valor que toma la variable dependiente cuando $x=0$.

* $\pmb{\beta_1}$ representa el **cambio** que se produce en la variable dependiente por cada unidad que **aumenta** la independiente.

Es justo aquí en donde aparece la **regresión lineal simple** como metodología para **ajustar** un modelo **lineal** que explique el comportamiento general de un conjunto de datos.

Supongamos ahora que encuestamos a **100** profesionales bogotanos y recolectamos información relacionada con el puntaje que obtuvieron en una prueba para el dominio de una segunda lengua, y la probabilidad que tienen de ser admitidos en un programa de postgrado. Podemos representar la información obtenida a través de un **gráfico de dispersión** como los que observas a continuación:

<figure>
<center><img src="fig2.png" width="500px" alt="Relación Lineal"></center>
</figure>

* En el gráfico **(a)** observamos una _relación lineal positiva_ entre las variables, lo que significaría que, por cada punto adicional en la prueba de segunda lengua, mayor es la probabilidad de admisión.

* En el gráfico **(b)** la _relación lineal_ entre las variables es _negativa_, lo que significaría que por cada punto adicional en la prueba de segunda lengua, la probabilidad de admisión disminuye.

* En el gráfico **(c)** observamos que _si existe_ una relación entre las variables, pero ésta _no es lineal_

* Finalmente, en el gráfico **(d)** vemos que, sin importar cuánto cambie el puntaje de la prueba de segundo idioma, la probabilidad de admisión no varía, y esto significa que _no existe relación alguna_ entre las variables.

Retomando nuestro ejemplo, vamos a suponer que el diagrama de dispersión para los datos recolectados es el siguiente

<figure>
<center><img src="fig3.png" width="500px" alt="Relación Lineal"></center>
</figure>

La recta que atraviesa la nube de puntos parece explicar el comportamiento de los datos, sin embargo **¿cuál es la recta que mejor los representa?** Para responder esta pregunta, debemos plantear primero el **modelo** que vamos a **ajustar** y que está dado por

<span style="color:#033D74">$$\pmb{y=\beta_0+\beta_1x+\epsilon}$$</span>

en donde el término $\pmb{\epsilon}$ se denomina el **error** del modelo y representa la _distancia vertical_ que existe entre los puntos y la recta. De esta manera, la **mejor recta de regresión** para modelar los datos, es aquella cuyos valores de $\beta_0$ y $\beta_1$ hacen que el **error** sea lo más pequeño posible, y está representada por la ecuación

<span style="color:#033D74">$$\pmb{\widehat{y}=\widehat{\beta}_0+\widehat{\beta}_1x}$$</span>

Es poco frecuente sin embargo, explicar el cambio que se produce sobre una variable dependiente a partir de una única covariable, y por ello, es necesario extender este tipo de relaciones al ámbito de la **regresión lineal múltiple**. Aquí, existe ahora un conjunto de variables independientes o covariables, que explican los cambios que sufre una variable respuesta, y el modelo está determinado por

<span style="color:#033D74">$$\pmb{y=\beta_0+\beta_1x_1+\beta_2x_2+\beta_3x_3+\cdots+\beta_kx_k+\epsilon}$$</span>

En el contexto de nuestro ejemplo, podemos pensar entonces en que, además de la calificación obtenida en una prueba específica para evaluar el dominio de una segunda lengua, el ranking de la universidad a la que se aspira ingresar y la divulgación científica también son variables que determinan la probabilidad de ser admitidos en un programa de postgrado. Nuevamente debemos ajustar un modelo que nos permita explicar adecuadamente la relación entre las variables y que, además, nos permita **predecir** o **pronosticar** la probabilidad de admisión.

Actualmente existen diversas metodologías para encontrar los valores de los coeficientes del modelo, entre otras, encontramos el _método de máxima verosimilitud_, el método del _gradiente descendente_, y, también el método de _**mínimos cuadrados ordinarios**_ que emplearemos en el ejemplo posterior. Todos ellos tienen sus ventajas y desventajas, y pueden usarse dependiendo del contexto, el recurso computacional, el número de parámetros, e incluso la naturaleza de los datos. 

Para revisar el funcionamiento de estas metodologías, podemos remitirnos por ejemplo a [https://towardsdatascience.com/linear-regression-simplified-ordinary-least-square-vs-gradient-descent-48145de2cf76](https://towardsdatascience.com/linear-regression-simplified-ordinary-least-square-vs-gradient-descent-48145de2cf76)


## <span style="color:#8F3A84">**Ejemplo**</span>
---

Continuando con la temática propuesta a lo largo del documento, implementaremos los algoritmos de **regresión lineal simple y múltiple** en un conjunto de datos que explica la probabilidad de admisión en distintas universidades para programas de maestría. 

Los datos fueron tomados de 

_**Mohan S Acharya, Asfia Armaan, Aneeta S Antony : A Comparison of Regression Models for Prediction of Graduate Admissions, IEEE International Conference on Computational Intelligence in Data Science 2019**_

y pueden ser descargados de [Kaggle](https://www.kaggle.com/mohansacharya/graduate-admissions).

La base de datos consta de $500$ registros y $8$ variables:

* GRE scores (GRE)

* TOEFL scores (TOEFL)

* University Ranking (URating)

* Statement of Purpose and Letter of Recommendation Strength (SOP y LOR)

* Undergraduate GPA  (CGPA)

* Research Experience (Research)

* Chance of Admit (Chance, variable respuesta)


```python
# Importamos librerías

import sklearn as skl
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

Tras importar las librerías necesarias para la implementación del proceso, cargamos la base de datos y visualizamos sus primeros registros:

```python
# Cargamos la base de datos y damos nombres a sus columnas
datos=pd.read_csv('/home/natalia/Descargas/Admission_Predict_Ver1.1.csv',header=0,sep=',')
datos.columns=['Serial','GRE','TOEFL','URating','SOP','LOR','CGPA','Research','Chance']
print('\nFilas x Columnas: ',datos.shape)
print(' ')
datos.head()
```

Para realizar un análisis adecuado, determinamos si la base de datos contiene datos **perdidos** o **faltantes** empleando un mapa de calor. Como se observa en el gráfico resultante, la base de datos carece de este tipo de datos.

```python
plt.figure(figsize=(12,4))
sns.heatmap(datos.isnull(),cbar=False,cmap='Spectral',yticklabels=False)
plt.title('Datos Faltantes en la Base',fontdict={'fontsize':18});
```

Empleando la librería **seaborn**, podemos representar las relaciones entre las variables empleando gráficos de dispersión, como se observa a continuación:

```python
sns.pairplot(datos,height=2.5)
plt.tight_layout()
plt.show()
```

A partir del gráfico, observamos por ejemplo, que la relación entre la probabilidad de admisión y el puntaje en la prueba GRE puede considerarse lineal. **¿Qué podríamos afirmar sobre la relación entre la variable respuesta y el rating de la universidad?**

Con la finalidad de analizar si las relaciones entre variables son de carácter lineal, podemos calcular los **coeficientes de correlación**, que toma valores entre $-1$ y $1$, entendiendo que, valores cercanos a **cero** indican una relación lineal débil o inexistente entre cada par de variables, mientras que, valores cercanos a $-1$ y $1$, significan que la relación lineal es fuerte.

Los coeficientes de correlación también pueden representarse en un mapa de calor como el siguiente:

```python
cor=datos.corr()
sns.set(font_scale=1.0)

plt.figure(figsize=(16, 6))
cm=sns.heatmap(cor,
            cmap='Spectral_r',
            cbar=True,
            annot=True,
            square=True,
            fmt='.2f',
            annot_kws={'size':10})
cm.set_title('Correlación', fontdict={'fontsize':18}, pad=12);
```

Vemos por ejemplo, como la relación entre las variables CGPA y Chance es fuerte y positiva, por lo que implementaremos una regresión lineal simple con ese par de variables.



## <span style="color:#8F3A84">**Regresión Lineal Simple: Chance vs. CGPA**</span>

---

Con la finalidad de evaluar el **rendimiento** de nuestro modelo, fraccionaremos el conjunto de datos aleatoriamente: **70%** de los datos serán empleados para entrenar el modelo, mientras que, el **30%** restante será empleado para probarlo:

```python
from sklearn.model_selection import train_test_split
x=datos['CGPA']
y=datos['Chance']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```


Ahora construimos un gráfico de dispersión con las variables seleccionadas y el fraccionamiento del conjunto de datos, en el que, como lo indica el coeficiente de correlación, puede sugerirse que un modelo lineal podría explicar la relación entre las dos variables.


```python
plt.figure(figsize=(10, 6))
plt.plot(x_train,y_train,'p',color='tab:green',label='Entrenamiento')
plt.plot(x_test,y_test,'d',color='tab:red', label='Prueba',alpha=0.5)
plt.title('Gráfico de Dispersión',fontdict={'fontsize':18})
plt.xlabel('CGPA')
plt.ylabel('Chance of Admit')
plt.legend(loc='upper left')
plt.show()
```

Una vez ajustemos el modelo, podemos calcular los valores ajustados de la variable Chance, y podemos probar el ajuste con los datos de prueba. Presentamos además algunas medidas que permiten evaluar el rendimiento del modelo:

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error,r2_score

X_train=np.array(x_train)
Y_train=np.array(y_train)# A continuación se ajusta el modelo de regresión lineal simple


modelo=linear_model.LinearRegression()
modelo.fit(X_train.reshape(-1,1),Y_train)

print('Intercepto: %.3f' % modelo.intercept_)
print('Pendiente: %.3f' % modelo.coef_[0])

Y_pred_train=modelo.predict(X_train.reshape(-1,1))

# Evaluación del Rendimiento
Y_test=np.array(y_test)
X_test=np.array(x_test)
Y_pred_test=modelo.predict(X_test.reshape(-1,1))


print('MSE Entrenamiento: %.3f, MSE Prueba: %.3f' % (
    mean_squared_error(Y_train,Y_pred_train),
    mean_squared_error(Y_test,Y_pred_test)))

print('R² Entrenamiento: %.3f, R² Prueba: %.3f' % (r2_score(Y_train,Y_pred_train),
                                                  r2_score(Y_test,Y_pred_test)))
```

El **MSE** _(Mean Squared Error)_ corresponde a la varianza de los errores del modelo, es decir, constituye un indicador de _**qué tan dispersos**_ están los datos al rededor de la recta ajustada, siendo deseables valores pequeños de la misma. 

De otra parte, el **Coeficiente de Determinación** o $\pmb{R²}$, es una medida de _bondad de ajuste_ para el modelo, y nos da información sobre la _**proporción de variabilidad**_ explicada por el mismo, por lo tanto, toma valores entre 0 y 1 siendo deseables los valores cercanos a 1.

Cabe resaltar, que al ser semejantes los valores del MSE y $R²$ para los datos de entrenamiento y de prueba, no hay sobreajuste en el modelo.

A continuación se presenta el gráfico de dispersión y la recta ajustada, así como la gráfica de los residuales:

```python
# Representar el modelo ajustado
plt.figure(figsize=(10, 6))
plt.plot(X_train,Y_pred_train,label='Y_pred_Entrenamiento')
plt.plot(X_test,Y_pred_test,linestyle=':',color='red',linewidth=4,label='Y_pred_Prueba')
plt.plot(X_train,Y_train,'d',color='green',alpha=0.45,label='Entrenamiento')
plt.plot(X_test,Y_test,'p',color='purple',alpha=0.45,label='Prueba')
plt.title('Gráfica de Dispersión',fontdict={'fontsize':18})
plt.xlabel('CGPA')
plt.ylabel('Chance of Admit')
plt.legend(loc='upper left')
plt.show()
```

```python
res_train=Y_pred_train-Y_train
res_train_stand=(res_train-res_train.mean())/res_train.std()

res_test=Y_pred_test-Y_test
res_test_stand=(res_test-res_test.mean())/res_test.std()

plt.figure(figsize=(10, 6))
plt.scatter(Y_pred_train,res_train_stand,marker='p',c='darkcyan',alpha=0.6,edgecolor='black',s=50.,label='Entrenamiento')
plt.scatter(Y_pred_test,res_test_stand,marker='d',c='mediumorchid',alpha=0.6,edgecolor='black',s=50.,label='Prueba')
plt.xlabel('Valores Ajustados')
plt.ylabel('Residuales')
plt.legend(loc='lower right')
plt.hlines(y=0,xmin=0.3,xmax=1.1,color='red',lw=2,ls='--')
plt.xlim([0.3,1.1])
plt.title('Ajustados vs. Residuales',fontdict={'fontsize':18})
plt.show()
```

## <span style="color:#8F3A84">**Regresión Lineal Múltiple**</span>

---

Ahora emplearemos todas las variables de la base de datos para explicar la probabilidad de ser admitido a un programa de maestría:

```python
X=datos.iloc[:,1:-1]
X.head()
```

Nuevamente repartiremos nuestro conjunto de datos para entrenamiento (70%) y prueba (30%) para evaluar el rendimiento del modelo:

```python
x_train_m,x_test_m,y_train_m,y_test_m=train_test_split(X,y,test_size=0.3)
X_train_m=np.array(x_train_m)
Y_train_m=np.array(y_train_m)

modelo2=linear_model.LinearRegression()
modelo2.fit(X_train_m,Y_train_m)
print('B0: %.3f' % modelo2.intercept_)
print('B1: %.3f' % modelo2.coef_[0])
print('B2: %.3f' % modelo2.coef_[1])
print('B3: %.3f' % modelo2.coef_[2])
print('B4: %.3f' % modelo2.coef_[3])
print('B5: %.3f' % modelo2.coef_[4])
print('B6: %.3f' % modelo2.coef_[5])
print('B7: %.3f' % modelo2.coef_[6])
```

```python
X_test_m=np.array(x_test_m)
Y_test_m=np.array(y_test_m)
Y_train_pred_m=modelo2.predict(X_train_m)
Y_test_pred_m=modelo2.predict(X_test_m)
print('MSE Entrenamiento: %.3f, MSE Prueba: %.3f' % (
    mean_squared_error(Y_train_m,Y_train_pred_m),
    mean_squared_error(Y_test_m,Y_test_pred_m)))

print('R² Entrenamiento: %.3f, R² Prueba: %.3f' % (r2_score(Y_train_m,Y_train_pred_m),
                                                  r2_score(Y_test_m,Y_test_pred_m)))
```

Se observa que en el modelo ajustado no hay sobreajuste y que se explica con él más del 80% de la variabilidad de los datos. Además, cabe resaltar que la variabilidad de las observaciones con relación al ajuste es baja, hechos que se verifican con los gráficos siguientes:

```python
res_train_m=Y_train_pred_m-Y_train_m
res_train_stand_m=(res_train_m-res_train_m.mean())/res_train_m.std()

res_test_m=Y_test_pred_m-Y_test_m
res_test_stand_m=(res_test_m-res_test_m.mean())/res_test_m.std()

plt.figure(figsize=(10, 6))
plt.scatter(Y_train_pred_m,res_train_stand_m,marker='p',c='darkcyan',alpha=0.6,edgecolor='black',s=50.,label='Entrenamiento')
plt.scatter(Y_test_pred_m,res_test_stand_m,marker='d',c='mediumorchid',alpha=0.6,edgecolor='black',s=50.,label='Prueba')
plt.xlabel('Valores Ajustados')
plt.ylabel('Residuales')
plt.legend(loc='lower right')
plt.hlines(y=0,xmin=0.3,xmax=1.1,color='red',lw=2,ls='--')
plt.xlim([0.3,1.1])
plt.title('Ajustados vs. Residuales',fontdict={'fontsize':18})
plt.show()
```
