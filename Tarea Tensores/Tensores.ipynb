## **lgebra de Tensores**
### Ahora veremos como efectuar algunas operaciones algebr谩icas entre tensores

```python
# Construimos dos tensores a y b de la misma dimensi贸n (2,3):
a=tf.constant([[1,3,5],[2,4,6]])
b=tf.constant([[1,2,3],[4,5,6]])

print(a)
print('\n',b)
```

```python
# Suma de tensores (se realiza elemento a elemento):
suma=tf.add(a,b)
print(suma)

# O alternativamente
print('\n', a+b)
```

```python
# Producto de tensores componente a componente:
prodc=tf.multiply(a,b)
print(prodc)

# O tambi茅n
print('\n', a*b)
```

```python
# Producto de tensores matricialmente: No se puede efectuar entre a y b pues no son conformables!!!

# Se generan dos tensores conformables para ilustrar la operaci贸n:
a1=tf.constant([[2,3],[4,8]])
b1=tf.constant([[1,2],[3,5]])
prodt=tf.matmul(a1,b1)
print(a1)
print('\n',b1)
print('\n',prodt)

# Tambi茅n podemos hacer
print('\n', a1@b1)
```

## **Funciones de Reducci贸n**
### Podemos extraer alguna informaci贸n sobre los elementos de los tensores

```python
# Construimos un tensor de dimensi贸n 2-3 para ilustrar:
x=tf.constant([[1.0,13.0],[5.0,7.0]])
print(x)

# Identificamos el elemento mayor dentro del tensor:
print('\nELEMENTO MAYOR:',tf.reduce_max(x))
# Identificamos el elemento menor dentro del tensor:
print('\nELEMENTO MENOR:',tf.reduce_min(x))
# Calculamos el promedio de los componentes del tensor:
print('\nPROMEDIO DE LOS ELEMENTOS:',tf.reduce_mean(x))
# Identificamos la posici贸n del elemento mayor dentro del tensor:
print('\nPOSICIN ELEMENTO MAYOR:',tf.argmax(x))
# Aplica la funci贸n de activaci贸n SOFTMAX a los componentes del tensor exp(e[i,j])/sum(exp(e[i,]))
print('\nSOFTMAX:\n',tf.nn.softmax(x))
# Se suman las probabilidades obtenidas con la funci贸n SOFTMAX para verificar que da 1:
print('\nSUMA PROBABILIDADES SOFTMAX:',tf.reduce_sum(tf.nn.softmax(x),axis=1),"\n")
```

## **Indexaci贸n**
